```
from detecto import core, utils, visualize
 
dataset = core.Dataset('/content/drive/My Drive/data')
model = core.Model(['gun'])
 
model.fit(dataset)
```

```
image = utils.read_image('/content/drive/My Drive/data/images (7).jpeg')
predictions = model.predict(image)
 
 
 
 
# predictions format: (labels, boxes, scores)
 
labels, boxes, scores = predictions
 
# ['alien', 'bat', 'bat']
print(labels) 
 
#           xmin       ymin       xmax       ymax
# tensor([[ 569.2125,  203.6702, 1003.4383,  658.1044],
#         [ 276.2478,  144.0074,  579.6044,  508.7444],
#         [ 277.2929,  162.6719,  627.9399,  51841]])
print(boxes)
 
# tensor([0.9952, 0.9837, 0.5153])
print(scores)
```

```
visualize.show_labeled_image(image, boxes[3], labels[3])
```
